```python
import argparse
from label_studio.load_segmentation import load_json
from label_studio.convert_points import convet_points
from image_processing.load_image import load_image,load_image_as_numpy_array
from image_processing.crop_image import crop_image_segment
from image_processing.masking import apply_light_masking
from image_processing.composite_image import composite_image
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
input_image_path="./data/template1/template-1.jpg"
pattern_image_path="./data/template1/pattern6.jpg"
input_json_path="./data/template1/template-1.json"
output_image_path="./data/template1/template-1-result.jpg"

def main():
    print("Step 1 : Load Segments")
    segments_points=load_json(input_json_path)
    print("Step 2 : Load pattern")
    pattern_image=load_image_as_numpy_array(pattern_image_path)
    print("Step 3 : Load Image")
    base_image=load_image(input_image_path)
    print("Step 4 : Convert points")
    pixel_segments_points=convet_points(segments_points,base_image.size)
    print("Step 5 : Apply segment operations")
    for segment in pixel_segments_points:
        print("Step 5.1 : Crop Image")
        croped_image=crop_image_segment(base_image=base_image,
                                        pixel_segment=segment)
        
        print(type(croped_image))
        
        print("Step 5.1 : Apply light mask")
        light_mask_image=apply_light_masking(croped_image=croped_image,
                                             pattern_image=pattern_image,
                                             base_image=base_image)
        
        print("Step 5.2 : Composite Image")
        base_image=composite_image(segment=segment,
                                   base_image=base_image,
                                   croped_image=light_mask_image)
        
        base_image.save(output_image_path)
        plt.imshow(base_image)
main()
```

```python
from google.cloud import storage
def download_file_from_gcs(gcs_path,folder):
    storage_client = storage.Client()

    bucket = storage_client.bucket("mrdarshan-public")

    blob = bucket.blob(gcs_path.replace("https://storage.googleapis.com/mrdarshan-public/",""))
    blob.download_to_filename(folder+"/"+gcs_path.split("/")[-1])
```
